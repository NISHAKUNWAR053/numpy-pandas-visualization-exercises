2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):



```python
x = np.linspace(0, 10, 100)
y1 = np.sqrt(x)

```

```python

plt.plot(x, y1, label="y = .sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Square Root Function")
plt.legend()
```

```python
y2 = x ** 3
```

```python

plt.plot(x, y2, label="y = x**3")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic Function")
plt.show()
```

```python
y3 = 2 * x
```

```python
plt.plot(x, y3, label="y = 2x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Function")
plt.legend()
```

```python
y4 = 1 / (x + 1)
```

```python
plt.plot(x, y4, label="y = 1 / (x + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Rational Function")
plt.legend()
```

Combine the figures you created in the last step into one large figure with 4 subplots.



```python
x = np.linspace(0, 11, 100)
plt.figure(figsize=(20, 4))

#define my first subplot
plt.subplot(1,4,1)
plt.plot(x, y1, label="y = .sqrt(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Square Root Function")


#define my second subplot
plt.subplot(1,4,2)
plt.plot(x, y2, label="y = x**3")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic Function")


#define my third subplot
plt.subplot(1,4,3)
plt.plot(x, y3, label="y = 2x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Function")


#define my fourth subplot
plt.subplot(1,4,4)
plt.plot(x, y4, label="y = 1 / (x + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Rational Function")


```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.plot(x, y1, c='#30a2da', label='$\sqrt{x}$', alpha=1)
plt.plot(x, y2, c='orange', label='$x^{3}$', alpha=1)
plt.plot(x, y3, c= (1,0,1), label='$2^{x}$', alpha=1)
plt.plot(x, y4, c= '#444444', label='$1/(x+1)$', alpha=1)

# Adjust xlim and ylim
plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('Combined Charts')
plt.legend()

plt.show()
```
